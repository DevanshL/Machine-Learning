#### SIMPLE LINEAR REGRESSION FROM SCRATCH

```python
class SimpleRegression:
    def __init__(self):
        self.slope = None
        self.intercept = None
    
    def fit(self,x,y):
        mu_x = np.mean(x)
        mu_y = np.mean(y)
        numerator = np.sum((x-mu_x)*(y-mu_y))
        denominator = np.sum((x-mu_x)**2)
        self.slope = numerator/denominator
        self.intercept = mu_y - (self.slope*mu_x)
        
    def predict(self,x):
        if self.slope is None or self.intercept is None:
            raise RuntimeError('Model not fitted\n')
        return self.slope*x + self.intercept   # mx+c    


model = SimpleRegression()
model.fit(x,y)

prediction = model.predict(x)
```

```python
print("Slope (m):", model.slope)
print("Intercept (b):", model.intercept)
print("Predictions:", prediction)
```
